```python
import pandas as pd

import numpy as np

df = pd.read_csv("/content/drive/MyDrive/Titanic_Dataset.csv")

df
```

```python
rows, columns = df.shape
print(f"Rows: {rows}, Columns: {columns}")
```

```python
df.drop("Name",axis=1,inplace=True)

df.drop("Ticket",axis=1,inplace=True)

df.drop("PassengerId",axis=1,inplace=True)

df.drop("Cabin",axis=1,inplace=True)

df.drop("Embarked",axis=1,inplace=True)
```

```python
updated_df = df
updated_df['Age']=updated_df['Age'].fillna(updated_df['Age'].mean())
updated_df.info()
```

```python
#2.1.3 Deleting the columns with missing data
updated_df = df.dropna(axis=1)
updated_df.info()
```

```python
  #2.2.1 Discretization

# Importing pandas and numpy libraries
import pandas as pd
import numpy as np

df = pd.DataFrame({'Age': [42, 15, 67, 55, 1, 29, 75, 89, 4, 10, 15, 38, 22, 77]})

df['Label'] = pd.cut(x=df['Age'], bins=[0, 3, 17, 63, 99],labels=['Baby/Toddler', 'Child', 'Adult','Elderly'])

print(df)

# Check the number of values in each bin
print("Categories: ")
print(df['Label'].value_counts())
```
