```python
from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# ---------------- BFS ----------------
def bfs(start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        print(node, end=" ")

        # (A) write the line that marks this node as visited
        visited.add(node)

        for nei in graph[node]:
            # (B) write the condition that prevents revisiting nodes
            if nei not in visited:
                # (C) write the line that adds neighbour to queue
                queue.append(nei)


# ---------------- DFS ----------------
def dfs(node, visited=None):
    if visited is None:
        visited = set()

    print(node, end=" ")

    # (D) write the line to mark node visited
    visited.add(node)

    for nei in graph[node]:
        # (E) write the condition for DFS recursion
        if nei not in visited:
            # (F) write the recursive DFS call
            dfs(nei, visited)


# ---------------- IDDFS ----------------
def dls(node, depth):
    if depth == 0:
        print(node, end=" ")
        return

    for nei in graph[node]:
        # (G) write the recursive call to DLS with depth-1
        dls(nei, depth - 1)


def iddfs(start, max_depth):
    for d in range(max_depth + 1):
        print(f"\nDepth {d}: ", end="")
        dls(start, d)

print("\nBFS:")
bfs('A')
print("\nDFS:")
dfs('A')
print("\n\nIDDFS:")
iddfs('A', 2)
```
